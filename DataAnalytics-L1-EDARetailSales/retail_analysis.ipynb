```python

new_tp=pd.merge(df_trans,df_prod,on='ProductID')
new_tp.groupby('ProductName')['Quantity'].sum()
```

```python
top_10_data=new_tp.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False).head(10)
top_10_data.plot(kind='bar',title='Top 10 Products',ylabel='Quantity',color='green')
#sns.barplot(x='ProductName',y='Quantity',data=new_tp).sort_values(ascending=False).head(10)
#
```

Road Clothing and How Vegetables display the highest contribution in top-selling products, indicating that these two specific items drive the highest overall sales volume for the business.

```python
new_tp['Revenue']=new_tp['Quantity']*new_tp['UnitPrice']
rev_data=new_tp.groupby('Category')['Revenue'].sum()
rev_data.plot(kind='bar',ylabel='Revenue',rot=0)
```

The Electronics category has the highest contribution In terms of revenue generation as compared to groceries

My 3 business insights:

1.Strategic Bundling Over Discounts: Our correlation matrix showed no relationship between discount size and purchase volume. Instead of price-dropping low-performing groceries, we should bundle them with our top-selling Electronics to drive traffic.

2.Youth Market Expansion: Customer age distribution is relatively flat. We have a massive opportunity to capture a larger share of the younger demographic by introducing tech-forward loyalty programs or targeted social media campaigns.

3.Optimizing Female Apparel Sales: While male shoppers currently drive a slightly higher sales volume, we can balance the demographic by directly targeting female shoppers with tailored marketing for our already high-performing "Clothing" category.
